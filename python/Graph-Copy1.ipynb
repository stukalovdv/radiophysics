```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
```

```python
%matplotlib inline
```

```python
my_dpi = 90
```

```python
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams["figure.figsize"] = (1920/my_dpi,1080/my_dpi)
plt.rcParams["lines.linewidth"] = 1
plt.rcParams["axes.prop_cycle"] = cycler(color=['r', 'g', 'b', 'y'])
plt.style.use("seaborn-whitegrid")
```

```python
fig, axes = plt.subplots(3,3)

data_tags = pd.read_csv('main_data.dat',sep='\t',header=0,index_col=False)
data = np.loadtxt("main_data.dat", skiprows = 1)

axes[0, 0].set(title="$E_{r}$",xlabel="$r$",ylabel="$E_{r}$")
axes[0, 1].set(title="$E_{\\varphi}$",xlabel="$r$",ylabel="$E_{\\varphi}$")
axes[0, 2].set(title="$E_{z}$",xlabel="$r$",ylabel="$E_{z}$")
axes[1, 0].set(title="$H_{r}$",xlabel="$r$",ylabel="$H_{r}$")
axes[1, 1].set(title="$H_{\\varphi}$",xlabel="$r$",ylabel="$H_{\\varphi}$")
axes[1, 2].set(title="$H_{z}$",xlabel="$r$",ylabel="$H_{z}$")
axes[2, 0].set(title="$j_{r}$",xlabel="$r$",ylabel="$j_{r}$")
axes[2, 1].set(title="$j_{\\varphi}$",xlabel="$r$",ylabel="$j_{\\varphi}$")
axes[2, 2].set(title="$j_{z}$",xlabel="$r$",ylabel="$j_{z}$")

axes[0, 0].plot(data[:,0],data[:,1])
axes[0, 1].plot(data[:,0],data[:,2])
axes[0, 2].plot(data[:,0],data[:,3])
axes[1, 0].plot(data[:,0],data[:,4])
axes[1, 1].plot(data[:,0],data[:,5])
axes[1, 2].plot(data[:,0],data[:,6])
axes[2, 0].plot(data[:,0],data[:,7])
axes[2, 1].plot(data[:,0],data[:,8])
axes[2, 2].plot(data[:,0],data[:,9])

plt.subplots_adjust(wspace=0.15, hspace=0.35)
plt.show()
fig.savefig('test.jpg')
```
