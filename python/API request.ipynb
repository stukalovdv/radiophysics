```python
import requests
import json
```

```python
payload = {"name":"User"}
response = requests.get("https://data.sensor.community/airrohr/v1/filter/box=56.388259,43.740941,56.219625,44.088700",params = payload)
```

```python
all_data = json.loads(response.text)
print("Найдено ",len(all_data)," датчиков!\n")
print(type(all_data[0]),'\n')
itterator = 0
for sens_num in range(len(all_data)):
    if all_data[sens_num]['sensordatavalues'][0]["value_type"] == "P1":
        sub_data = all_data[sens_num]['sensordatavalues']
        itterator += 1
        print("Запись №",itterator,sub_data[0]['value_type'],":",sub_data[0]['value'],sub_data[1]['value_type'],":",sub_data[1]['value'],'\n')
```
