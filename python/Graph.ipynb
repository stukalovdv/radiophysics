```python
#fig, axes = plt.subplots(5,3)
PATH = "DataByTime.dat"
fig = plt.figure()

data_tags = pd.read_csv(PATH,sep='\t',header=0,index_col=False)
data = np.loadtxt(PATH, skiprows = 1)
limit_x = data[len(data[:,0]) - 1,0]

ax_Er_t = fig.add_subplot(5, 3, 1, title = '$E_{r}(t)$', xlabel = "$t$", ylabel = "$E_{r}$")
ax_Er_t.plot(data[:,0],data[:,1])
ax_Er_t.set_xlim([0, limit_x])

ax_Ep_t = fig.add_subplot(5, 3, 2, title = '$E_{\\varphi}(t)$', xlabel = "$t$", ylabel = "$E_{\\varphi}$")
ax_Ep_t.plot(data[:,0],data[:,2])
ax_Ep_t.set_xlim([0, limit_x])

ax_Ez_t = fig.add_subplot(5, 3, 3, title = '$E_{z}(t)$', xlabel = "$t$", ylabel = "$E_{z}$")
ax_Ez_t.plot(data[:,0],data[:,3])
ax_Ez_t.set_xlim([0, limit_x])

ax_Hr_t = fig.add_subplot(5, 3, 4, title = '$H_{r}(t)$', xlabel = "$t$", ylabel = "$H_{r}$")
ax_Hr_t.plot(data[:,0],data[:,4])
ax_Hr_t.set_xlim([0, limit_x])

ax_Hp_t = fig.add_subplot(5, 3, 5, title = '$H_{\\varphi}(t)$', xlabel = "$t$", ylabel = "$H_{\\varphi}$")
ax_Hp_t.plot(data[:,0],data[:,5])
ax_Hp_t.set_xlim([0, limit_x])

ax_Hz_t = fig.add_subplot(5, 3, 6, title = '$H_{z}(t)$', xlabel = "$t$", ylabel = "$H_{z}$")
ax_Hz_t.plot(data[:,0],data[:,6])
ax_Hz_t.set_xlim([0, limit_x])

ax_Jr_t = fig.add_subplot(5, 3, 7, title = '$J_{r}(t)$', xlabel = "$t$", ylabel = "$J_{r}$")
ax_Jr_t.plot(data[:,0],data[:,7])
ax_Jr_t.set_xlim([0, limit_x])

ax_Jp_t = fig.add_subplot(5, 3, 8, title = '$J_{\\varphi}(t)$', xlabel = "$t$", ylabel = "$J_{\\varphi}$")
ax_Jp_t.plot(data[:,0],data[:,8])
ax_Jp_t.set_xlim([0, limit_x])

ax_Jz_t = fig.add_subplot(5, 3, 9, title = '$J_{z}(t)$', xlabel = "$t$", ylabel = "$J_{z}$")
ax_Jz_t.plot(data[:,0],data[:,9])
ax_Jz_t.set_xlim([0, limit_x])

ax_Hz_t = fig.add_subplot(5, 1, 4, title = '$H_{z}(t)$', xlabel = "$t$", ylabel = "$H_{z}$")
ax_Hz_t.plot(data[:,0],data[:,6])
ax_Hz_t.set_xlim([0, limit_x])

Spec = np.fft.rfft(data[:,6],16*len(data[:,6]))
Spec = np.abs(Spec)
Omega_spec_x = np.linspace(0,len(Spec),len(Spec))

ax_Spec_omega = fig.add_subplot(5, 1, 5, title = '$S(\\omega)$', xlabel = "$\\omega$", ylabel = "$S$")
ax_Spec_omega.plot(Omega_spec_x/(16*3.14),Spec[:len(Spec)])
ax_Spec_omega.set_xlim([0, 6])

print(len(Omega_spec_x))
#ax_Spec_omega.set_xlim([0, 10])
plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
fig.savefig('TimePic.jpg')
```
